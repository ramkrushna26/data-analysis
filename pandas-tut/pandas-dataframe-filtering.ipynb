```python
import pandas as pd
```

```python
df = pd.read_csv("employees.csv", parse_dates = ["HIRE_DATE"])
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].fillna(0.0).astype(int)
df['MANAGER_ID'] = df['MANAGER_ID'].fillna(0.0).astype(int)
df["JOB_ID"] = df["JOB_ID"].astype('category')
```
